```python
df['window'] = df['date'].dt.floor('h')
```

```python
grouped=df.groupby(['window','user'])
```

```python
df['is_logon']  = (df['activity'] == 'Logon').astype(int)
df['is_logoff'] = (df['activity'] == 'Logoff').astype(int)
```

```python
features = df.groupby(['user','window']).agg(
    logon_count = ('is_logon','sum'),
    logoff_count = ('is_logoff','sum'),
    unique_pcs = ('pc','nunique')
).reset_index()
```

```python
user_stats = features.groupby('user').agg(
    avg_logon = ('logon_count','mean'),
    std_logon = ('logon_count','std'),
    avg_pcs   = ('unique_pcs','mean'),
    std_pcs=('unique_pcs','std')
)
user_stats
```

```python
user_stats.describe()
```

```python
user_stats['avg_pcs'].max()
```

```python
features.head()
```

```python
features['hour']=features['window'].dt.hour
```

```python
features.groupby('hour')['logon_count'].mean()
#checking the mean at each hour in the day by all users
```

```python
enriched_df=pd.merge(features,user_stats,on='user')
enriched_df
```

```python
enriched_df['z_logon']=(enriched_df['logon_count']-enriched_df['avg_logon'])/enriched_df['std_logon']
enriched_df['z_pcs']=(enriched_df['unique_pcs']-enriched_df['avg_pcs'])/(enriched_df['std_pcs']+0.01)
```

```python
enriched_df.head()
```

```python
enriched_df['night_flag']=(enriched_df['hour']<6)|(enriched_df['hour']>22)
```

```python
enriched_df[enriched_df['night_flag']==True]
```

```python
enriched_df['behavior_score']= 0.4* abs(enriched_df['z_logon']) + 0.35*abs(enriched_df['z_pcs']) + 0.25*enriched_df['night_flag']
```
