```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
```

```python
df=pd.read_csv("enriched_logon.csv")
df.head()
```

```python
df['behavior_score'].describe()
```

```python
df.info()
```

```python
X = df[[
  'logon_count',
  'logoff_count',
  'unique_pcs',
  'hour',
  'z_logon',
  'z_pcs',
  'night_flag'
]]
```

```python
X.head()
```

```python
std=StandardScaler()
```

```python
df_std=std.fit_transform(X)
```

```python
model=IsolationForest()
```

```python
pred=model.fit_predict(df_std)
```

```python
df['behavior_score']
```

```python
sample=df.copy()
```

```python
sample['pred']=pred
```

```python
sample[['behavior_score', 'pred']].describe()
```

```python
sample[sample['pred'] == -1][['behavior_score', 'pred']]
```

```python
sample[sample['pred'] == 1][['behavior_score', 'pred']]
```

```python
sample[sample['behavior_score']==sample['behavior_score'].max()]
```
